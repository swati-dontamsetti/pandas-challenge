```python
# Dependencies and Setup
import pandas as pd

# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#print data frame
school_data_complete.head()
```

```python
# calculate length of array of unique schools and all students
diffSchools = len(school_data_complete["school_name"].unique())
sumStudents = len(school_data_complete["student_name"])

# calculate various data points
totBudget = school_data_complete["budget"].unique().sum()
avgMath = school_data_complete["math_score"].mean()
avgReading = school_data_complete["reading_score"].mean()

# this is what the directions say the over passing rate should be
# avgOverall = (avgMath + avgReading)/2

# create an array of the require data points, then calculate percentage
passingMath = school_data_complete.loc[school_data_complete["math_score"] >= 70, :]
percentMath = (len(passingMath)/sumStudents)*100
passingReading = school_data_complete.loc[school_data_complete["reading_score"] >= 70, :]
percentReading = (len(passingReading)/sumStudents)*100

# the actual overall passing rate should be the average of the % passing math and reading
avgOverall = (percentMath + percentReading)/2

# create data frame with information
frame_district = pd.DataFrame({
    "Total Schools": diffSchools,
    "Total Students": sumStudents,
    "Total Budget": totBudget,
    "Average Math Score": avgMath,
    "Average Reading Score": avgReading,
    "% Passing Math": percentMath,
    "% Passing Reading": percentReading,
    "% Overall Passing Rate": [avgOverall]
})

# format the data frame
frame_district["Total Students"] = frame_district["Total Students"].map("{:,}".format)
frame_district["Total Budget"] = frame_district["Total Budget"].map("${:,.2f}".format)
frame_district["Average Math Score"] = frame_district["Average Math Score"].map("{:.2f}".format)
frame_district["Average Reading Score"] = frame_district["Average Reading Score"].map("{:.2f}".format)
frame_district["% Passing Math"] = frame_district["% Passing Math"].map("{:.2f}%".format)
frame_district["% Passing Reading"] = frame_district["% Passing Reading"].map("{:.2f}%".format)
frame_district["% Overall Passing Rate"] = frame_district["% Overall Passing Rate"].map("{:.2f}%".format)

#print data frame
frame_district
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# print the min and max of per student budget in order to create a range
print(school_summary["Per Student Budget"].min())
print(school_summary["Per Student Budget"].max())

# create bins and names
bins = [0, 580, 610, 640, 670]
group_names = ["<$580", "$580-610", "$610-640", "$640-670"]
```

```python
# add a new column to the original data set
school_data_complete["Spending Ranges (Per Student)"] = pd.cut((school_data_complete["budget"]/school_data_complete["size"]),bins,labels=group_names)

# groupby the spending ranges
per_student = school_data_complete.groupby("Spending Ranges (Per Student)")

# calculate the new info
avg_m = per_student["math_score"].mean()
avg_r = per_student["reading_score"].mean()
math_pass = school_data_complete[school_data_complete["math_score"] >= 70].groupby("Spending Ranges (Per Student)")["student_name"].count()/per_student["student_name"].count()*100
read_pass = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("Spending Ranges (Per Student)")["student_name"].count()/per_student["student_name"].count()*100
overall_pass = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].groupby("Spending Ranges (Per Student)")["student_name"].count()/per_student["student_name"].count()*100

# create data frame
scores_by_spending = pd.DataFrame({
    "Average Math Score": avg_m,
    "Average Reading Score": avg_r,
    "% Passing Math": math_pass,
    "% Passing Reading": read_pass,
    "% Overall Passing Rate": overall_pass
})

# format
scores_by_spending["Average Math Score"] = scores_by_spending["Average Math Score"].map("{:.2f}".format)
scores_by_spending["Average Reading Score"] = scores_by_spending["Average Reading Score"].map("{:.2f}".format)
scores_by_spending["% Passing Math"] = scores_by_spending["% Passing Math"].map("{:.2f}%".format)
scores_by_spending["% Passing Reading"] = scores_by_spending["% Passing Reading"].map("{:.2f}%".format)
scores_by_spending["% Overall Passing Rate"] = scores_by_spending["% Overall Passing Rate"].map("{:.2f}%".format)

# print
scores_by_spending
```
